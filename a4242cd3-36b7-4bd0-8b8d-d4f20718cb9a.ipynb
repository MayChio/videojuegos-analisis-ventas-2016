```python
ventas_por_plataforma = df.groupby('platform')['total_sales'].sum()

# Identificar las plataformas con las mayores ventas totales
plataformas_top_ventas = ventas_por_plataforma.nlargest(5).index  

# Filtrar el DataFrame original para incluir solo las plataformas seleccionadas
df_top_plataformas = df[df['platform'].isin(plataformas_top_ventas)]

# Construir una distribución basada en los datos de cada año para las plataformas seleccionadas
distribucion_por_año = {}
for plataforma in plataformas_top_ventas:
    df_plataforma = df_top_plataformas[df_top_plataformas['platform'] == plataforma]
    ventas_por_año = df_plataforma.groupby('year_of_release')['total_sales'].sum()
    distribucion_por_año[plataforma] = ventas_por_año

# Identificar plataformas que solían ser populares pero que ahora no tienen ventas
plataformas_antiguas = ventas_por_plataforma[ventas_por_plataforma == 0].index

# Mostrar los resultados
print("Ventas por Año para Plataformas con Mayores Ventas Totales:")
for plataforma, ventas_por_año in distribucion_por_año.items():
    print(f"Plataforma: {plataforma}")
    print(ventas_por_año)
    print()

print("Plataformas que solían ser populares pero que ahora no tienen ventas:")
print(plataformas_antiguas)
```

```python
#Calcularo el promedio directo

primer_año = df.groupby('platform')['year_of_release_int'].min()
ultimo_año = df.groupby('platform')['year_of_release_int'].max()
# Calcular la duración de vida de cada plataforma
duracion_de_vida = ultimo_año - primer_año

# Calcular los promedios directos
duracion_promedio_nuevas = (duracion_de_vida == 0).mean()
duracion_promedio_antiguas = (duracion_de_vida > 0).mean()

print("Duración promedio de vida de las plataformas nuevas:", duracion_promedio_nuevas)
print("Duración promedio de vida de las plataformas antiguas:", duracion_promedio_antiguas)
```

        Hipótesis nula (H0): Las calificaciones promedio de los usuarios para Xbox y PC son iguales..

```python
# Filtrar los puntajes de usuario para las plataformas Xbox y PC
xbox_scores = df[df['platform'] == 'XOne']['user_score']
pc_scores = df[df['platform'] == 'PC']['user_score']

# Realizar el test de hipótesis (t-test)
t_statistic, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=False)

# Imprimir el resultado del test de hipótesis
print("Resultado del test de hipótesis:")
print(f"Estadístico t: {t_statistic}")
print(f"Valor p: {p_value}")

# Interpretar los resultados
alpha = 0.05
if p_value < alpha:
    print("Conclusión: Rechazamos la hipótesis nula. Las calificaciones promedio de los usuarios para Xbox y PC son diferentes.")
else:
    print("Conclusión: No podemos rechazar la hipótesis nula. No hay suficiente evidencia para concluir que las calificaciones promedio de los usuarios para Xbox y PC son diferentes.")
```
