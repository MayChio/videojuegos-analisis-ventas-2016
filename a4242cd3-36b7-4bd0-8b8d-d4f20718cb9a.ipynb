```python
ventas_por_plataforma = df.groupby('platform')['total_sales'].sum()

# Identificar las plataformas con las mayores ventas totales
plataformas_top_ventas = ventas_por_plataforma.nlargest(5).index  

# Filtrar el DataFrame original para incluir solo las plataformas seleccionadas
df_top_plataformas = df[df['platform'].isin(plataformas_top_ventas)]

# Construir una distribución basada en los datos de cada año para las plataformas seleccionadas
distribucion_por_año = {}
for plataforma in plataformas_top_ventas:
    df_plataforma = df_top_plataformas[df_top_plataformas['platform'] == plataforma]
    ventas_por_año = df_plataforma.groupby('year_of_release')['total_sales'].sum()
    distribucion_por_año[plataforma] = ventas_por_año

# Identificar plataformas que solían ser populares pero que ahora no tienen ventas
plataformas_antiguas = ventas_por_plataforma[ventas_por_plataforma == 0].index

# Mostrar los resultados
print("Ventas por Año para Plataformas con Mayores Ventas Totales:")
for plataforma, ventas_por_año in distribucion_por_año.items():
    print(f"Plataforma: {plataforma}")
    print(ventas_por_año)
    print()

print("Plataformas que solían ser populares pero que ahora no tienen ventas:")
print(plataformas_antiguas)
```
